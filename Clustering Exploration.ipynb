## Part1: K Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, silhouette_score
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from tqdm import tqdm
from sklearn.decomposition import PCA
```

```python
def totalCostForElbow(data,model):
    cost=0
    for i in range(data.shape[0]):
        point = data[i]
        centroid = model.cluster_centers_[model.labels_[i]]
        cost += np.sum(np.square(point-centroid))
    return cost
```

```python
#implement elbow method from scratch
def elbow(X,maxK):
    costArr=[]
    kArr=[]
    for k in tqdm(range(1,maxK)):
        model = KMeans(n_clusters=k,n_init=30)
        model.fit(X)
        cost = totalCostForElbow(X,model)
        costArr.append(cost)
        kArr.append(k)
    plt.plot(kArr, costArr, 'bo-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Cost')
    plt.title('The Elbow Method using WCSS')
    plt.show()
```

```python
#implement silhouette method from scratch
def silhouette(X,maxK):
    sKArr=[]
    silhoutteScores=[]
    for k in tqdm(range(2,maxK)):
        model = KMeans(n_clusters=k)
        model.fit(X)
        cluster_labels = model.labels_
        sKArr.append(k)
        silhoutteScores.append(silhouette_score(X, cluster_labels))
    plt.plot(sKArr,silhoutteScores,'bo-')
    plt.xlabel('Number of Clusters') 
    plt.ylabel('Silhouette score') 
    plt.title('Silhouette analysis For Optimal k')
    plt.show()
```

## Part2: Hierarchical clustering

```python
# implement hierarchical clustering ( you can use scikit module)
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import sys

def hierarchical_clustering(X,n,typ):
    pca = PCA(n_components = 2)
    reducedX = pca.fit_transform(X)
    singleModel = AgglomerativeClustering(n_clusters = n,linkage=typ)
    plt.figure(figsize =(12, 12))
    plt.scatter(reducedX[:,0], reducedX[:,1],c = singleModel.fit_predict(reducedX), cmap ='rainbow')
    plt.title(typ)
    plt.show()
    return
```

```python
#visualize a dendogram
def dendogram(scaledX,typ):
    plt.figure(figsize =(8, 8))
    plt.title('Visualising the data')
    if(typ=='single'):
        rd = sys.getrecursionlimit()
        sys.setrecursionlimit(10**7)
        Dendrogram = sch.dendrogram((sch.linkage(scaledX, method=typ)))
        sys.setrecursionlimit(rd)
    else:
    Dendrogram = sch.dendrogram((sch.linkage(scaledX, method=typ)))
    return
```

```python
dendogram(scaledX,'single')
```

```python
dendogram(scaledX,'ward')
```

**Question**: Explain the procedure to find the optimal number of clusters using dendogram

Answer: We need to find the maximum vertical line which is not cut by any horizontal line if extended. Then if we draw a horizontal line cutting that vertical line , then the number of vertical lines it cuts are the number of clusters.

Here the optimal number of clusters seems to be 3.





### Below taking clusters as 4 so as to plot the graph

```python

hierarchical_clustering(scaledX,4,'single')
hierarchical_clustering(scaledX,4,'complete')
hierarchical_clustering(scaledX,4,'ward')
```

**Question:** Mention the differences observed using single and complete linkage:

Answer: Due to chaining effect, in single linkage , clusters get combined with each other because of min dist being paramters of clubing. Therefore we see clusters of large size. 
On the other hand, Complete linkage gives us compact and more spherical clusters because clusters do not get clubbed easily due to maximum distance beign the parameter

## Part3: Image Compression

```python
import cv2
# Boiler plate code
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# cv2_imshow(img)
img_size = img.shape

# Reshape it to be 2-dimension
X = img.reshape(img_size[0] * img_size[1], img_size[2])
```

```python
kValues=[2,5,10,20,50,75,100]
for k in tqdm(range(len(kValues))):
    model = KMeans(n_clusters=kValues[k])
    model.fit(X)
    newX=[]
    for i in range(X.shape[0]):
        label = model.labels_[i]
        newX.append(list(model.cluster_centers_[label]))
    newX= np.asarray(newX)
    compressedImage = np.clip(newX.astype('uint8'),0,255)
    ogn_size = img_size[0]*img_size[1]*img_size[2]*8
    c_size = img_size[0]*img_size[1]*img_size[2]*(np.log2(kValues[k])) + kValues[k]*24
    image = compressedImage.reshape(img_size[0] , img_size[1],3)

    print("Original size: ",ogn_size/(8*1024),"KB")
    print("Compressed Size: ",c_size/(8*1024),"KB")
    print("Reduced Size =",(ogn_size-c_size)/(8*1024),"KB")
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
    plt.title('Number of clusters: '+str(kValues[k]))
    plt.show()
```
